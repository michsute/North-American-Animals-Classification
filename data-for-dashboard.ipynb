```python
# add random timestamp
import random
from datetime import datetime, timedelta

coordinates = {
    'San Juan Mntns, Colorado': (37.5, -107.5),
    'Lebec, California': (34.8616, -118.8695),
    'Archbold, FL': (27.6114, -81.5352),
    None: (39.5501, -105.7821)  # NaN wird mit Koordinaten von Colorado ersetzt
}


# Funktion, um einen zufälligen Timestamp im Jahr 2024 zu generieren
def random_timestamp():
    start_date = datetime(2024, 1, 1)  # Startdatum 1. Januar 2024
    end_date = datetime(2024, 12, 31, 23, 59, 59)  # Enddatum 31. Dezember 2024, 23:59:59
    delta = end_date - start_date
    random_seconds = random.randint(0, int(delta.total_seconds()))  # Zufällige Anzahl von Sekunden
    return start_date + timedelta(seconds=random_seconds)

# Funktion, um Koordinaten basierend auf der Location zu bekommen
def get_coordinates(location):
    if location in coordinates:
        return coordinates[location]
    else:
        return None, None  # Falls Ort nicht im Dictionary gefunden wird

# Füge die 'Latitude' und 'Longitude' Spalten hinzu
df[['latitude', 'longitude']] = df['location'].apply(lambda loc: pd.Series(get_coordinates(loc)))


# Füge eine neue Spalte 'Timestamp' hinzu, die für jede Zeile einen zufälligen Timestamp enthält
df['timestamp'] = [random_timestamp() for _ in range(len(df))]
```

```python
# Speichern des DataFrames als CSV
df.to_csv('./streamlit/test-dataset/df_dashboard-02-time-long-lat.csv', index=False)
```
